```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\hp\OneDrive\Documents\Investment.csv")
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
x=pd.get_dummies(x,dtype='int')
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=9)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_pred=model.predict(x_test)
```

# MSE

```python
from sklearn.metrics import mean_squared_error
print("MSE for test data",mean_squared_error(y_test,y_pred))
```

# RMSE

```python
print("RMSE for test score",np.sqrt(mean_squared_error(y_test,y_pred)))
```

# ACCURACY

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score, cross_validate
score = cross_val_score(model, x, y, cv=5)
print(score)
print("Cross Validation Score:", score.mean())
```

# checking the feature elimination

```python
x = np.append(arr = np.ones((50,1)).astype(int), values = x, axis = 1) 
```

```python
import statsmodels.api as sm
x_opt = x[:,[ 0,1,2,3,4,5]]  # Select columns by position
model_ols = sm.OLS(endog=y, exog=sm.add_constant(x_opt)).fit()
model_ols.summary()
```
