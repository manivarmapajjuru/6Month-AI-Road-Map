```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\hp\Downloads\emp_sal.csv")
```

```python
df
```

```python
x=df.iloc[:,1:2]
x.head(3)
```

```python
y=df.iloc[:,2]

y.head(3)
```

# linear_modelm

```python
from sklearn.linear_model import LinearRegression 
model_LR=LinearRegression()
model_LR.fit(x,y)
```

```python
# linear Plot Visualization
plt.scatter(x,y,c='r')
plt.plot(x,model_LR.predict(x),c='b')
plt.title("LinearRegression Visulization")
plt.xlabel("Poisition_level")
plt.ylabel("Salary")
```

```python
model_LR.predict([[6]])
```

# Polynominal Regression 

#### where independent x variable increases the power by default it takes degree 2
###### like quadaratic_expression etc

```python
from sklearn.preprocessing  import PolynomialFeatures
model_POLY = PolynomialFeatures()
x_poly=model_POLY.fit_transform(x)  # to change the x values to power
model_POLY.fit(x_poly,y)
model_LR.fit(x_poly,y)
```

```python
predict=model_LR.predict(model_POLY.fit_transform([[6]]))
predict
```

```python
x.shape
```

```python
# visualiztion for ploy
```

```python
plt.scatter(x,y,c='r')
plt.plot(x,model_LR.predict(model_POLY.fit_transform(x)),c='b')
plt.title("Polynomial_Regression")
plt.xlabel("Poisition_Level")
plt.ylabel("Salary")
plt.show()
```

##### degree 5

```python
from sklearn.preprocessing  import PolynomialFeatures
model_POLY = PolynomialFeatures(degree=5)
x_poly=model_POLY.fit_transform(x)  # to change the x values to power
model_POLY.fit(x_poly,y)
model_LR.fit(x_poly,y)
predict=model_LR.predict(model_POLY.fit_transform([[6]]))
print(predict)
plt.scatter(x,y,c='r')
plt.plot(x,model_LR.predict(model_POLY.fit_transform(x)),c='b')
plt.title("Polynomial_Regression")
plt.xlabel("Poisition_Level")
plt.ylabel("Salary")
plt.show()
```

# SVM

```python
from sklearn.svm import SVR
model_SVR=SVR(kernel='poly')
model_SVR.fit(x,y)
model_SVR.predict([[6]])
```

```python
from sklearn.svm import SVR
model_SVR=SVR(kernel='poly',degree=4,gamma="scale",C=1.0)
model_SVR.fit(x,y)
model_SVR.predict([[6]])
```

# K-nearest neighbors

```python
from sklearn.neighbours import KNeighbourRegressor 
model_KNN=KNeighbourRegressor(n_neighbour=5)
model_KNN.fit(x,y)
model_KNN.predict([[6]])
```
