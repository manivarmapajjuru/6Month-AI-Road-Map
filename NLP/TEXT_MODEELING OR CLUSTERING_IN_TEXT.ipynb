# Text Preprocessing

##### remove punctuation 
##### remove stopword
##### stemming/lemmization

```python
from nltk.stem import WordNetLemmatizer
wnl=WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
corpus=[]
for i in range(len(df_200)):
    rp=re.sub('[^a-zA-Z]'," ",df_200['Article'][i])
   # print(rp)
    rp=rp.split()
    #print(rp)
    rp=[wnl.lemmatize(word) for word in rp if not word in stopwords.words('english')]
    #print(rp)
    rp=" ".join(rp)
    #print(rp)
    corpus.append(rp)
   
```

# vectorization

```python
from sklearn .feature_extraction.text import CountVectorizer
cv= CountVectorizer()
x=cv.fit_transform(corpus)
```

# Modeeling using LDA

```python
from sklearn.decomposition import LatentDirichletAllocation
model=LatentDirichletAllocation(n_components=4)
model.fit(x)
```

```python
topic_result=model.transform(x)
```

# combining witn Original Data

```python
df_200 = df.head(200).copy()  # 🔐 Make an actual copy to avoid SettingWithCopyWarning

df_200['group'] = topic_result.argmax(axis=1)  # ✅ Safe assignment
```
