```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\hp\OneDrive\Documents\SRK_CLASS\SRK_Class\Notes\22. Natural Language Processing\SMSSpamCollection",sep="\t",names=['Label','Message'])
```

```python
import nltk
import re
from nltk.corpus import stopwords  # Corrected import
from nltk.stem import PorterStemmer

ps = PorterStemmer()
```

```python
corpus=[]
for i in range(len(df)):
    rp=re.sub('[^a-zA-Z]'," ",df['Message'][i])
   # print(rp)
    rp=rp.split()
    #print(rp)
    rp=[ps.stem(word) for word in rp if not word in stopwords.words('english')]
    #print(rp)
    rp=" ".join(rp)
    #print(rp)
    corpus.append(rp)
print(corpus)   
```

# vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x=cv.fit_transform(corpus).toarray()
x
```

```python
y=pd.get_dummies(df['Label'],drop_first=True).astype('int')
```

```python
y
```

 # NAVIE BAYES

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
model = MultinomialNB()
model.fit(x_train, y_train)
```

# prediction

```python
ypred_test=model.predict(x_test)
ypred_train=model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
print("Train Accuracy",accuracy_score(y_train,ypred_train))
print("Test Accuracy",accuracy_score(y_test,ypred_test))
```

# new data

```python
# New message input
new_msg = ["Congratulations! You won a brand new car. Call now!"]

# Preprocess new message same as corpus
new_corpus = []
for message in new_msg:
    rp = re.sub('[^a-zA-Z]', ' ', message)
    rp = rp.lower().split()
    rp = [ps.stem(word) for word in rp if word not in stopwords.words('english')]
    rp = " ".join(rp)
    new_corpus.append(rp)

# Vectorize new message using the same CountVectorizer
new_x = cv.transform(new_corpus).toarray()

# Predict
pred = model.predict(new_x)

# Show result
print("Prediction:", "Spam" if pred[0] == 1 else "Ham")
```
